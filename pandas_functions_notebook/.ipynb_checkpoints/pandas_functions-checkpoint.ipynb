```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                      'B': ['B0', 'B1', 'B2']},
                     index=['K0', 'K1', 'K2'])
 

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                       'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
 

result = left.join(right,how='outer')
result.head()
```

```python
result = left.join(right,how='outer')
result
```

```python
result = left.join(right,how='inner')
result
```

```python
result = left.join(right,how='right')
result
```

```python
result = left.join(right,how='left')
result
```

```python
result = pd.merge(left, right, left_index=True, right_index=True, how='inner')
result
```

```python
result = pd.merge(left, right,left_index=True, right_index=True)
result
```

```python
left.index.rename('index', inplace=True)
right.index.rename('index', inplace=True)
left
```

```python
result = pd.merge(left, right,on='index', how='inner')
result
```

```python
#map
```

```python
def gender_convert(g):
    if g=='Male':
        return 0
    elif g=='Female':
        return 1
```

```python
data.gender = data.gender.map(gender_convert)
```

```python
data.head()
```

```python
data.gender = data.gender.map({'Male':0,'Female':1})
data.head()
```

```python
data.gender = data.gender.apply(gender_convert)
data.head()
```

```python
#group by
```

```python
df = pd.DataFrame({'name':['ali','ali','ali','mahdi'],
                   'score':[10,12,9,15],
                   'rankk':[4,3,2,1]})
df
```

```python
df.groupby('name').max()
```

```python
df.groupby('name').score.mean()
```

```python
df.groupby('name').rankk.mean()
```
